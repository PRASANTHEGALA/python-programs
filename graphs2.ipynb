```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([10,5,8,8])
mylabels=(["Python","Maths","English","chemistry"])
d=([0.2,0,0,0])
plt.pie(ypoints,explode=d)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([10,4,8,9])
mylabels=["CSE","ECE","Civil","Mech"]
mycolors = ["black", "red", "green", "blue"]
plt.pie(xpoints,labels=mylabels, colors=mycolors)
plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np
xpoints=np.array(["English","Maths","chemistry","python"])
ypoints=np.array([5,6,7,8])
plt.title("Time table")

plt.xlabel("Hours")

plt.ylabel("Subjects")

plt.plot(xpoints,ypoints,ls='dotted',marker='o')
plt.show()
```
