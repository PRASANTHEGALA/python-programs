```python
#Bar graph
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["BMW", "MERCEDES-BENZ", "TOYOTA", "OTHER SUPER CARS"])
y = np.array([67, 76, 86, 24])
plt.xlabel("AUTOMOBILE BRANDS")
plt.ylabel("CUSTOMER CONSUMPTION RATE")
plt.bar(x,y,width=0.55,color="black")
plt.show()
```

```python
#Horizontal Bar graph
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["BMW", "MERCEDES-BENZ", "TOYOTA", "OTHER SUPER CARS"])
y = np.array([67, 76, 86, 24])
plt.xlabel("AUTOMOBILE BRANDS")
plt.ylabel("CUSTOMER CONSUMPTION RATE")
plt.barh(x,y,color="black")
plt.show()
```

```python
#Pie chart
import matplotlib.pyplot as plt
import numpy as np
y = np.array([451, 366, 254,125])
label = ["BMW", "TOYOTA", "MERCEDES-BENZ", "PEUGEOT"]
bdcolors = ["y", "black", "green", "c"]
myexplode = [0,0.2,0,0,2]
plt.pie(y,labels = label, colors = bdcolors,shadow=True,)
plt.legend()
plt.show() 
```

```python
#markers
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([92, 23, 66, 22])
plt.plot(ypoints, marker = '*')
plt.grid()
plt.show()
```

```python
#markers with size increase
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([92, 23, 66, 22])
plt.plot(ypoints, marker = 'o',ms=20)
plt.grid()
plt.show()
```

```python
#markers with point colour change
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([92, 23, 66, 22])
plt.plot(ypoints, marker = 'o',ms=20,mec='r')
plt.title("alcohol consumption in AP")
plt.xlabel("Average drinkers per city")
plt.ylabel("litres consumed")
plt.grid()
plt.show()
```
